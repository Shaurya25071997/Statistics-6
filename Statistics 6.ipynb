## Q1. What is a Cumulative Distribution Function, and how does it work?

The cumulative Distribution function calculates the cumulative probability for a given x-value. Use the CDF to determine the probabilty that a random observation that is taken from the population will be less than equal to certain value.You can also use this incformtaion to determine the probability that an observation will be greater than a certain value or between two values.

How CDF works we will understand it by a exmaple

**Example of using the  CDF to evaluate fill weights**

For example, cold drink can fill weights follow a normal distribution with a mean of 12 ounces and a standard deviation of 0.25 ounces. The probability density function (PDF) describes the likelihood of possible values of fill weight. The CDF provides the cumulative probability for each x-value.
